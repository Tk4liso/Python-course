**Práctica While**

**Objetivo**

En este caso práctico se propone al alumno la "Creación del software de un ascensor". Para ello, debe hacer uso de la sentencia de control de flujo While en Python.



Instrucciones

Implementa una función mover_ascensor que reciba dos argumentos: piso_actual y piso_deseado

Dentro de la función implementa un bucle while para mover el ascensor a cada piso, uno por uno.

El bucle while debe comprobar que el piso actual es distinto al deseado, para conocer si el usuario ha llegado.

Dentro del bucle while debe comprobarse si el piso actual es mayor o menor que el piso deseado, para conocer si el ascensor debe subir o bajar.

En caso de que el ascensor deba subir, se imprimirá por pantalla un mensaje que indique a que piso está subiendo el ascensor y en qué piso se encuentra en ese momento.

Por último, la variable piso_actual deberá incrementarse en 1.

En caso de que el ascensor deba bajar, se seguirá el proceso inverso con los mismos pasos que para la subida, restando 1 a la variable piso_actual.

Finalmente, al terminar el bucle while, se imprimirá un mensaje indicando que se ha llegado al piso deseado.

Invoca la función mover_ascensor proporcionándole un piso actual y un piso deseado.



Resultado esperado

Subiendo al piso 6. Piso actual: 2

Subiendo al piso 6. Piso actual: 3

Subiendo al piso 6. Piso actual: 4

Subiendo al piso 6. Piso actual: 5

Piso 6 alcanzado

```python
#Solución 1

def mover_ascensor(piso_actual, piso_deseado):
    while piso_actual != piso_deseado:
        if piso_actual > piso_deseado:
            print(f"Bajando al piso {piso_actual - 1}. Piso actual: {piso_actual}")
            piso_actual -= 1
        elif piso_actual < piso_deseado:
            print(f"Subiendo al piso {piso_actual + 1}. Piso actual: {piso_actual}")
            piso_actual += 1
    else:
        print(f"Piso {piso_actual} alcanzado.")

mover_ascensor(6,1)
```
