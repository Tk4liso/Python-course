````python
#SOLUCIÓN

def formato_notificacion(func):
    def wrapper(self, plataforma):
        print(f"***** NUEVO EVENTO en {plataforma} *****")
        func(self, plataforma)
        print(f"***** FIN DEL EVENTO en {plataforma} *****")
    return wrapper

class NotificadorRedSocial:
    
    def __init__(self):
        self._evento = "Lanzamiento de un nuevo libro"
    
    @formato_notificacion
    def notificar_evento(self, plataforma):
        if plataforma == "Twitter":
            print("Tweet:", self._evento)
        elif plataforma == "Instagram":
            print("Instagram Story:", self._evento)
        elif plataforma == "Facebook":
            print("Post en Facebook:", self._evento)
        else:
            print("Plataforma no soportada.")


notificador = NotificadorRedSocial()
plataformas = ["Twitter", "Instagram", "Facebook", "LinkedIn"]

for plataforma in plataformas:
    notificador.notificar_evento(plataforma)
````

# ¿Qué me faltó entender sobre los decorators?

## 1. El `wrapper` debe recibir los mismos argumentos que la función decorada

Mi `wrapper` estaba definido así:

```python
def wrapper():
```

Pero la función decorada era:

```python
def notificar_evento(self, plataforma):
```

Como el `wrapper` reemplaza a la función original, debe aceptar los mismos argumentos para poder recibirlos y pasarlos a la función decorada.

```python
def wrapper(self, plataforma):
    func(self, plataforma)
```

---

## 2. El decorator no tiene acceso automático a las variables de la función

En el `wrapper` intenté usar `plataforma` sin haberla recibido como parámetro.

Un decorator sólo puede utilizar la información que recibe mediante sus parámetros o la que ya existe en su propio alcance. Si necesita `plataforma`, debe recibirla:

```python
def wrapper(self, plataforma):
```

---

## 3. `func` representa la función original

En un decorator, el parámetro `func` es la función que está siendo decorada.

Por eso, dentro del `wrapper`, llamar a:

```python
func(self, plataforma)
```

equivale a ejecutar la función `notificar_evento`.

---

## 4. Olvidé ejecutar el método

Escribí:

```python
noti1.notificar_evento
```

Eso sólo obtiene una referencia al método.

Para ejecutarlo es necesario usar paréntesis y pasar los argumentos:

```python
noti1.notificar_evento("Twitter")
```

---

## Conclusión

Mi principal error fue no entender que **el `wrapper` sustituye a la función original**. Por eso debe recibir los mismos argumentos y, cuando sea necesario, llamar a la función original (`func`) pasándole esos argumentos. De esa forma, el decorator puede agregar comportamiento antes y después de ejecutar la función decorada.

### Práctica 2 - Decorators (Nivel Medio)

**Objetivo**

Reforzar el uso de decorators en métodos que reciben argumentos.

**Instrucciones**

1. Define un decorator llamado `registrar_transaccion`.

2. Antes de ejecutar la función decorada debe imprimir:

```text
--- Inicio de transacción ---
```

3. Después de ejecutarla debe imprimir:

```text
--- Fin de transacción ---
```

4. Crea una clase llamada `Banco`.

5. La clase debe tener un atributo `_saldo`, inicializado desde el constructor.

6. Agrega un método llamado `retirar(self, cantidad)` que imprima:

```text
Retirando $<cantidad>
```

> **No es necesario validar el saldo.** El objetivo es practicar decorators.

7. Aplica el decorator al método `retirar`.

8. Crea una instancia y realiza varios retiros con cantidades diferentes.

.

.

**Resultado esperado (ejemplo)**

```text
--- Inicio de transacción ---
Retirando $500
--- Fin de transacción ---
```

````python
def registrar_transaccion(func):

    def wrapper(self, cantidad):
        print("--- Inicio de transacción ---")
        func(self, cantidad)
        print("--- Fin de transacción ---")
    return wrapper

class Banco:

    def __init__(self, saldo):
        self._saldo = saldo

    @registrar_transaccion
    def retirar(self, cantidad):
        print("Retirando:", cantidad)
        self._saldo -= cantidad
        print("Saldo actual:", self._saldo)


# === MAIN ===

nose1 = Banco(500)
nose1.retirar(40)
````
