**Práctica Diccionarios**

**Objetivo**

Crear un sistema para la gestión del inventario de un negocio mediante Diccionarios y sus métodos en Python.



Instrucciones

Crea una clase llamada GestionInventario que administre el inventario de productos de una tienda.

La clase debe tener un método inicializador que defina un atributo de instancia inventario que inicialmente sea un diccionario vacío. Este atributo almacenará los productos como claves y las cantidades como valores.

Implementa un método añadir_producto(self, producto, cantidad) que añada un nuevo producto al inventario o actualice la cantidad si el producto ya existe.

Implementa un método eliminar_producto(self, producto) que utilice el método .pop() para eliminar un producto del inventario por su clave, si existe.

Implementa un método consultar_producto(self, producto) que utilice el método .get() para retornar la cantidad de un producto específico, o un mensaje indicando que el producto no existe: Producto no existe en el inventario.

Implementa un método mostrar_inventario(self) que imprima todos los productos y sus cantidades haciendo uso del método .items()  y un bucle for que los recorra para iterar sobre el diccionario e imprimir por pantalla el {producto}: {cantidad}.

Ejecución del programa:

Crea un objeto de la clase GestionInventario.

Llama 3 veces al método añadir_producto con 3 productos y cantidades diferentes. Por ejemplo: 10 Manzanas, 5 Peras y 5 Manzanas de nuevo (en este caso debería actualizar la cantidad total a 15).

Llama al método consultar_producto para consultar cuántas Manzanas hay e imprime por pantalla: Consultar Manzanas: {resultado de la función}.

Llama al método mostrar_inventario

Llama al método eliminar_producto, eliminando las Peras del inventario.

Llama al método de nuevo mostrar_inventario, imprimiendo antes la frase: Inventario después de eliminar Peras: {resultado de la función}.



Resultado esperado

Consultar Manzanas: 15

--- Inventario ---

Manzanas: 15

Peras: 5

------------------

Inventario después de eliminar Peras:

--- Inventario ---

Manzanas: 15

------------------

```python
#SOLUCIÓN 1

class GestionInventario:
    
    def __init__(self):
        self.inventario = {}
    
    def añadir_producto(self,producto,cantidad):
        if producto in self.inventario:
            self.inventario[producto] += cantidad
        else:
            self.inventario.update({producto:cantidad})
    
    def eliminar_producto(self,producto):
        if producto in self.inventario:
            self.inventario.pop(producto)
            print("Producto eliminado")
        else:
            print("El producto no existe en el inventario")
    
    def consultar_producto(self,producto):
        if producto in self.inventario:
            print(f"Consultar {producto}: {self.inventario.get(producto)}")
        else:
            print("Producto no existe en el inventario.")
    
    def mostrar_inventario(self):
        print("-" * 20)
        print("Producto | Cantidad")
        
        for clave, valor in self.inventario.items():
            print(f"{clave}: {valor}")



# === MAIN ===
objeto = GestionInventario()

objeto.añadir_producto("Manzanas",10)
objeto.añadir_producto("Peras",5)
objeto.añadir_producto("Manzanas",5)

objeto.consultar_producto("Manzanas")
objeto.mostrar_inventario()
objeto.eliminar_producto("Peras")

print("Inventario después de eliminar Peras:")
objeto.mostrar_inventario()
```

```python
#SOLUCIÓN 2

class GestionInventario:
    
    def __init__(self):
        self.inventario = {}
    
    def añadir_producto(self,producto,cantidad):
        if producto in self.inventario:
            self.inventario[producto] += cantidad
        else:
            self.inventario[producto] = cantidad
    
    def eliminar_producto(self,producto):
        if producto in self.inventario:
            self.inventario.pop(producto, None)
            print("Producto eliminado")
        else:
            print("El producto no existe en el inventario")
    
    def consultar_producto(self,producto):
        return self.inventario.get(producto, "Producto no existe en el inventario.")
    
    def mostrar_inventario(self):
        print("-" * 20)
        print("Producto | Cantidad")
        
        for producto, cantidad in self.inventario.items():
            print(f"{producto}: {cantidad}")



# === MAIN ===
objeto = GestionInventario()

objeto.añadir_producto("Manzanas",10)
objeto.añadir_producto("Peras",5)
objeto.añadir_producto("Manzanas",5)

print("Consultar Manzanas:", objeto.consultar_producto('Manzanas'))
objeto.mostrar_inventario()
objeto.eliminar_producto("Peras")

print("Inventario después de eliminar Peras:")
objeto.mostrar_inventario()
```
