### 1. Operaciones con Sets

```python
miset1 = {"azul", "rojo", "verde"}
```

```python
miset2 = {"morado", "rojo", "amarillo"}
```

```python
miset1.intersection(miset2)
```

```python
#Aparecen los elementos únicos, es decir, sin elementos repetidos
miset1.union(miset2)
```

```python
miset1
```

```python
miset2
```

```python
list(miset1.union(miset2))
```

```python
miset1.issubset(miset2)
```
