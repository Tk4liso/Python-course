Instrucciones

Crea una clase llamada Libro que represente los detalles básicos de un libro en una biblioteca. La clase Libro debe incluir los siguientes atributos de instancia:

titulo (str): El título del libro.

autor (str): El nombre del autor del libro.

isbn (str): El número ISBN del libro.

disponible (bool): Un valor booleano que indica si el libro está disponible para préstamo.

Dentro de la clase Libro, define un método __init__ que inicialice los atributos de instancia con los valores proporcionados como argumentos. Asegúrate de que todos los libros se inicialicen como disponibles.

Define un método información que muestre por pantalla los detalles de libro.

Define un método prestar_libro que cambie el estado de disponible a False cuando un libro se preste. Debe mostrarse por pantalla que el libro ha sido prestado y su información invocando el método información. Si el libro ya está prestado, debe imprimir un mensaje que diga: Este libro ya está prestado.

Define otro método devolver_libro que cambie el estado de disponible a True cuando un libro sea devuelto. Debe mostrarse por pantalla que el libro ha sido devuelto y su información invocando el método informacion. Si el libro ya está disponible, debe imprimir un mensaje que diga: Este libro ya está en la biblioteca

Fuera de la clase, crea al menos dos instancias de Libro con diferentes detalles. Ejemplo: un libro de J.K. Rowling y otro de George Orwell.

Prueba los métodos prestar_libro y devolver_libro con tus instancias de libros para demostrar que el estado de disponibilidad cambia apropiadamente y que los mensajes se imprimen cuando es necesario.

```python
#SOLUCION 1

class Libro():
    
    def __init__(self, titulo, autor, isbn):
        self._titulo = titulo
        self._autor = autor
        self._isbn = isbn
        self._disponible = True
    
    def informacion(self):
        if self._disponible:
            flag = "Disponible"
        else:
            flag = "No disponible"
        print("Título:", self._titulo, "\nAutor:", self._autor, "\nISBN:", self._isbn, "\nEn estantería:", flag)
    
    def prestar_libro(self):
        if self._disponible:
            self._disponible = False
            print("Libro prestado con éxito.")
            self.informacion()
        else:
            print("Este libro ya está prestado.")

    def devolver_libro(self):
        if self._disponible == False:
            self._disponible = True
            print(f"El libro {self._titulo} ha sido devuelto.")
        else:
            print("Este libro ya está en la biblioteca.")


# === MAIN ===

libro1 = Libro("Luna de Pluton", "Dross", 69)
libro2 = Libro("Meditaciones", "Marco Aurelio", 18)

libro1.informacion()
libro1.prestar_libro()
#libro1.informacion()
libro1.devolver_libro()
libro1.informacion()

print("-" * 30)

libro2.informacion()
libro2.prestar_libro()
#libro2.informacion()
libro2.devolver_libro()
libro2.informacion()
```
