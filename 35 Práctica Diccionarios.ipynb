**Objetivo**

Instrucciones

El programa tendrá una función llamada agregar_experiencia que recibe dos argumentos: perfil_laboral y nueva_experiencia.

perfil_laboral será un diccionario que contiene información personal del usuario: nombre, apellido, edad, ciudad, experiencias. Este último valor debe corresponderse con una lista que contendrá las experiencias laborales.

nueva_experiencia será una variable tipo string que indica el nuevo puesto de trabajo.

Dentro de la función agregar_experiencia, añade nueva_experiencia a la lista de experiencias en perfil_laboral y retorna el perfil_laboral como resultado de la función.

Crea otra función llamada generar_cv_reducido que tome perfil_laboral como argumento e imprima por pantalla un resumen del CV (con el formato especificado en las secciones siguientes), incluyendo la información personal y las experiencias laborales.

Fuera de las funciones añade una nueva experiencia laboral a tu perfil utilizando la función agregar_experiencia y llama a la función generar_cv_reducido tomando como argumento el resultado de la función agregar_experiencia para visualizar tu CV.



Resultado esperado

Edad: 30, Ciudad: Madrid

Experiencia: ['Ingeniera de software en XYZ Corp', 'Gerente de proyecto en ABC Inc']
