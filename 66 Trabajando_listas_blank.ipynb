### 1. Modificando una lista

```python
lista
```

Uno de los métodos de listas más utilizados es el método .append(), el cual agrega algo al final de la lista. Cabe recalcar que no sucede como en los strings, que se crea una copia del objeto. En este caso particular el método actúa directamente sobre nuestro objeto/variable y no sobre una copia del mismo.

```python
lista.append("amarillo")
```

```python
lista
```

```python
lista.append(["morado", "verde"])
```

```python
lista
```

Como se ve, el método .append() sirve para **añadir un único valor a nuestra lista.**

```python
lista = ["azul", "verda", "rojo"]
```

```python
lista.extend(["morado","amarillo","azul"])
```

```python
lista
```

El método .extend() es similar a .append() pero este nos permite concatenar los elementos de una lista a nuestra lista, en lugar de agregarlo como una sublista (como lo hace .append()).

```python
lista.remove("azul")
```

```python
lista
```

Como vemos, con el método .remove() se eliminó el primer "azul" de la lista, porque empieza a buscar de izquierda a derecha.

El método .reverse() le da la vuelta a la lista original.

```python
lista.reverse()
```

```python
lista
```

```python
lista_reversa = lista.reverse()
```

```python
lista_reversa
```

```python
type(lista_reversa)
```

En la variable lista_reversa no hay nada porque la transformación del método .reverse() se está aplicando sobre el objeto/variable lista. Esto mismo sucede con los demás métodos de esta sección.

```python
lista
```

El método .pop() nos va a devolver el último elemento de nuestra lista para que podamos asignárselo a una variable. 

Nota: además de devolverlo como un return también lo elimina de la lista original.

```python
lista.pop()
```

```python
lista
```

```python
lista2 = lista
```

```python
lista2
```

```python
lista[0] = "rojo"
```

```python
lista
```

```python
lista2
```

La nueva variable lista2 también cambia de valor luego de que lista[0]="rojo" porque cuando nosotros hacemos lista2 = lista, estamos haciendo una referencia al valor de lista.

Cuando hacemos lista2 = lista no estamos haciendo una copia de *lista*, estamos haciendo una referencia de lista2 hacia el mismo objeto que apunta la variable *lista*, que es concretamente esta lista:

['verda', 'rojo', 'morado', 'amarillo']

Entonces, cuando cambiamos el objeto *lista* a través de alguna de sus variables que la están referenciando también van a camiar **todas** las variables que lo referencien. Porque cambiamos el objeto original (el valor que se referencia) y por ende también cambia el valor el todas las variables que referencien ese objeto.

Entonces, para hacer de verdad una copia de una lista y no una referencia a un mismo valor hacemos uso del método **.copy()**.

```python
lista2 = lista.copy()
```

```python
lista
```

```python
lista2
```

```python
lista[0] = "mayonesa"
```

```python
lista
```

```python
lista2
```

El método .clear() "limpia" la lista y la deja como una lista vacía.

```python
lista2.clear()
```

```python
lista2
```

### 2. Accediendo a elementos de la lista

```python
lista = ["azul","verde","rojo"]
```

```python
lista.index("verde")
```

```python
lista.index("rojo")
```

```python
lista.extend(["azul","azul"])
```

```python
lista
```

Hay que tener cuidado con el método .index() porque si hay elementos repetidos nos va a mostrar el índice del primer elemento (más cercano a la izquierda).

```python
lista.index("azul")
```

El método .count() nos muestra el número de elementos duplicados.

```python
lista.count("azul")
```
