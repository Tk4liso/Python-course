**Objetivo**

Crear un simulador de cuota mensual de préstamos hipotecarios haciendo uso de objetos y sus métodos en python.



Instrucciones

Crea una clase llamada SimuladorPrestamo que modele el proceso de calcular el pago total de un préstamo hipotecario, incluyendo tanto el capital inicial como los intereses generados a lo largo del plazo del préstamo.

La clase toma como parámetros detalles_prestamo, un número complejo cuya parte real será la entrada proporcionada para la hipoteca y la parte imaginaria el tipo de interés.

También tomará como parámetro opcional los años del préstamos hipotecario (por defecto 30) y el precio de la vivienda (por defecto para este ejercicio 300000).

Deben almacenarse en el método constructor de la clase al menos los siguientes atributos de instancia:

tasa_mensual: Tasa de interés (parte imaginaria de detalles_prestamo) dividida entre 12 para convertirla a mensual y dividida entre 100.

numero_pagos: Numero de pagos totales (número de años * 12).

entrada: Parte real del parámetro detalles_prestamo.

prestamo: Precio de la vivienda restando la entrada.

cuota_mensual: Parámetro a calcular en el programa (inicialmente 0).

Implementa un método llamado calcular_pago_total que realice el cálculo de la cuota mensual a pagar del préstamo hipotecario y que actualice el atributo cuota_mensual. El cálculo de la cuota mensual se realiza mediante la siguiente fórmula:

Cuota mensual = prestamo * (tasa de interés mensual * (1 + tasa de interés mensual)**número de pagos) / ((1 + tasa de interés mensual)**snúmero de pagos - 1)

Implementa otro método mostrar_resultados que imprima la simulación de la hipoteca: el precio de la vivienda, la entrada, los años y finalmente la cuota mensual.

Instancia la clase con un número complejo representativo y ejecuta los métodos para calcular el pago total y mostrar los resultados.



Resultado esperado

----- Simulación Hipoteca -----

Para una vivienda de 300000 euros, aportando una entrada de 50000.0 euros y con una tasa de interés del 2.5% anual durante 30 años:

Cuota mensual a pagar: 987.8022470442972 euros

----- Fin de la Simulación -----

```python
class SimuladorPrestamo:
    
    def __init__(self, detalles_prestamo, años_prestamo=30, precio_vivienda=300000):
        self.años_prestamo = años_prestamo
        self.precio_vivienda = precio_vivienda
        
        self.entrada = detalles_prestamo.real
        self.tasa_mensual = (detalles_prestamo.imag / 12) / 100
        
        self.numero_pagos = años_prestamo * 12
        self.prestamo = precio_vivienda - self.entrada
        self.cuota_mensual = 0

    def calcular_pago_total(self):
        self.cuota_mensual = self.prestamo * (self.tasa_mensual * (1 + self.tasa_mensual)**self.numero_pagos) / ((1 + self.tasa_mensual)**self.numero_pagos - 1)
    
    def mostrar_resultados(self):
        print("----- Simulación Hipoteca -----")
        
        print(f"\nPrecio de la vivienda: {self.precio_vivienda}")
        print(f"\nEntrada: {self.entrada}")
        print(f"\nAños: {self.años_prestamo}")
        print(f"\nCuota mensual: {self.cuota_mensual}")


# === MAIN ===
hipoteca = SimuladorPrestamo(3+5j)
hipoteca.calcular_pago_total()
hipoteca.mostrar_resultados()
```
