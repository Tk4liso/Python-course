**Objetivo**

Desarrollar un programa que asista en la planificación y diseño de proyectos arquitectónicos mediante el cálculo de áreas y volúmenes de diversas estructuras utilizando la importación de módulos en Python.



Instrucciones

Módulo calculos_arquitectonicos.py:

Este módulo contendrá varias funciones útiles para el diseño arquitectónico:

area_rectangulo(largo, ancho): Calcula y retorna el área de un rectángulo.

volumen_cilindro(radio, altura): Calcula y retorna el volumen de un cilindro.

area_circulo(radio): Calcula y retorna el área de un círculo.

Nota: Para el cálculo de áreas y volúmenes puedes importar el modulo math que permite el uso de variables como el número pi.

Módulo main.py:

Importa las funciones area_rectangulo, volumen_cilindro, y area_circulo del módulo calculos_arquitectonicos.

Define una función llamada planificar_proyecto que realice las siguientes acciones:

Utiliza area_rectangulo para calcular el área destinada a la sala principal de una casa. Imprime por pantalla el resultado: Área de la sala principal: X metros cuadrados.

Emplea volumen_cilindro para determinar el volumen necesario para una piscina cilíndrica de diseño. Imprime por pantalla el resultado: volumen de la piscina: X metros cúbicos.

Aplica area_circulo para planificar un jardín circular alrededor de la propiedad. Imprime por pantalla el resultado: Área del jardín circular: X metros cuadrados.

Llama a la función planificar_proyecto para realizar todos los cálculos. Puedes tomar los siguiente parámetros como ejemplo:

Sala principal: 20 largo, 15 ancho

Piscina: 5 radio, 2 altura.

Jardín: 10 radio.



Resultado esperado

---------- Informe proyecto ----------

Volumen de la piscina: 157.07963267948966 metros cúbicos.
