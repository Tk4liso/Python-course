El método .pop() devuelve el valor de la clave especificada y a la vez elimina esa clave (y su valor) del diccionario. Similar a como funciona con las listas.

```python
diccionario.pop("key1")
```

```python
diccionario
```

```python
diccionario2 = diccionario.copy()
```

### 2. Accediendo a los elementos del diccionario

```python
diccionario = {
    "key1": 1,
    "key2": 2,
    "key3": 3
}
```

.values() permite acceder a los valores de las claves sin tener que dar el nombre de las claves.

```python
diccionario.values()

#Se puede convertir en una lista de la sig. forma:
list(diccionario.values())
```

.keys() da el nombre de las claves sin los valores

```python
diccionario.keys()

#Transformarlo en una lista
list(diccionario.keys())
```

**.items()** nos devuelve los valores y las claves como tuplas. Sirve para cuando vamos a iterar en un bucle for.

```python
diccionario.items()
```

```python
for key, value in diccionario.items():
    print(key)
    print(value)
    #print("{} -> {}".format(key, value))
```
