**Práctica Variables**

Objetivo

Crear un "Menú del Día de un Restaurante Gourmet" usando variables y la función print para mostrar los platos en la consola.



Instrucciones:

Escribe un programa en Python que utilice variables para almacenar información sobre diferentes platos y luego usa la función print para mostrar un menú gourmet.

El menú debe contener tres secciones: Entrante, Plato Principal y Postre.

Para cada sección, define dos variables:

Una para el nombre del plato: nombre_entrante, nombre_principal y nombre_postre.

Otra para el precio: precio_entrante, precio_principal y precio_postre.

Usa múltiples llamadas a la función print para mostrar primero el nombre del plato y luego su precio en líneas separadas.

Importante: Las variables que definas para crear el menú tienen que tener el mismo nombre que se muestra en el punto 3: nombre_entrante, precio_entrante, nombre_principal, precio_principal, nombre_postre y precio_postre.



Resultado esperado

Menú del Día - Restaurante Gourmet

-- Entrante --

Sopa de trufas

Precio:

12 euros

-- Plato Principal --

Filete Mignon con salsa de vino tinto

Precio:

30 euros

-- Postre --

Tiramisú casero

Precio:

8 euros



```python
#Nombre y precio de platillos
nombre_entrante = "Sopa de trufas"
precio_entrante = 12

nombre_principal = "Filete Mignon con salsa de vino tinto"
precio_principal = 30

nombre_postre = "Tiramisú casero"
precio_postre = 8

#Prints del menú
print("Menú del Día - Restaurante Gourmet")
print("--Entrante--")
print(nombre_entrante,"Precio: ", precio_entrante, " euros")
print("--Plato Principal--")
print(nombre_principal,"Precio: ", precio_principal, " euros")
print("--Postre--")
print(nombre_postre,"Precio: ", precio_postre, " euros")
```
