**Práctica If-else**

**Objetivo**

Programa un "Creador de Recomendaciones de Películas" utilizando la estructura de control de flujo if-else en Python.



Instrucciones

El programa incluirá una función llamada recomendar_pelicula que toma dos argumentos: genero y edad.

La función utilizará sentencias if/elif/else para recomendar una película basada en el género y la edad proporcionados.

Si el género es "acción" y la edad es mayor o igual a 13, recomendará "Deadpool".

Si el género es "acción" y la edad es menor de 13, recomendará "Regreso al futuro".

Si el género es "comedia", independientemente de la edad, recomendará "Aterriza como puedas".

Si el género no es "acción" ni "comedia", recomendará "Explorar otros géneros".

La función retornará el nombre de la película recomendada.

Fuera de la función, se pedirá al usuario que ingrese su género favorito y su edad, y se llamará a la función recomendar_pelicula con estos datos.



Resultado esperado

Teniendo en cuenta tu edad (26) y tu género favorito (acción), te recomiendo la siguiente película: Deadpool

```python
def recomendar_pelicula(genero, edad):
    if genero == "accion" and edad >= 13:
        return "Deadpool"
    elif genero == "accion":
        return "Regreso al futuro"
    elif genero == "comedia":
        return "Aterriza como puedas"
    else:
        return "Explorar otros géneros"

genero = input("¿Cuál es tu genero favorito?")
edad = int(input("¿Qué edad tienes?"))

print(f"Teniendo en cuenta tu edad ({edad}) y tu género favorito ({genero}), te recomiendo la siguiente película:\n{recomendar_pelicula(genero, edad)}")
```
