**Práctica Bytes y Bytearray**

**Objetivo**

Implementa un "Codificador de comunicaciones" en Python utilizando Bytes.



Instrucciones

El programa tendrá una función llamada crear_codigo_secreto que recibe un argumento: mensaje.

mensaje debe corresponderse con una cadena de bytes.

El objetivo de esta función es codificar el mensaje para que sea complicado de interpretar sin conocer el tratamiento que sea ha realizado.

Dentro de la función, realiza las siguientes operaciones:

Almacena en una variable repr_hex los bytes que ocupen una posición par en la cadena y transfórmalos a hexadecimal de la siguiente forma: repr_hex = repr_hex.hex()

Almacena en una variable repr_bin los bytes que ocupen una posición par en la cadena.

Extrae el primer y último byte de repr_bin en dos variables repr_bin_primero y repr_bin_ultimo. Transforma estos bytes a binario utilizando la función bin()

Convierte en String repr_hex, repr_bin, repr_bin_primero y repr_bin_ultimo y concaténalos.

Devuelve el resultado de la concatenación

Fuera de la función, imprime el mensaje sin codificar y el resultado codificado de la función crear_codigo_secreto.



Resultado esperado

Mensaje original: b'Este es un mensaje secreto'

Mensaje codificado: 45742073752065736a20657274b'Et su esj ert'0b10001010b1110100

```python
def crear_codigo_secreto(mensaje):
    #Extraer bytes en posiciones pares y convertirlos a hexadecimal
    repr_hex = mensaje[::2]
    repr_hex = repr_hex.hex()

    #Exxtraer bytes en posiciones pares
    repr_bin = mensaje[::2]

    
    repr_bin_primero = bin(repr_bin[0])
    repr_bin_ultimo = bin(repr_bin[-1])
    
    #return print(type(repr_hex), type(str(repr_bin)), type(repr_bin_primero), type(repr_bin_ultimo))

    return repr_hex + str(repr_bin) + repr_bin_primero + repr_bin_ultimo


# = Main =

mensaje = b'Hola, soy Goku'
mensaje_codificado = crear_codigo_secreto(mensaje)

print(f"Mensaje original: b'{mensaje}'")
print(f"Mensaje codificado:\n{mensaje_codificado}")
```
