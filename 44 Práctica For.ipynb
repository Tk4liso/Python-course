**Práctica For**

**Objetivo**

Programa una aplicación "TO-DO List" que recoja la lista de tareas que tienes pendientes de realizar. Para ello, utiliza la estructura de control de flujo for en Python.



Instrucciones

El programa incluirá una función llamada añadir_tarea

Esta función recibirá como argumento la tarea a añadir en formato string.

Dentro de la función se creará una variable lista tareas que debe contener la lista de tareas incluidas en el archivo tareas.txt.

Tras leer la lista de tareas del archivo tareas.txt, debe añadirse al final de la lista tareas la nueva tarea que se proporciona como argumento a la función.

Por último, la función retornará la variable tareas

El programa incluirá una función llamada gestionar_tareas que recibirá como argumento la lista de tareas retornada por la función anterior.

Mediante un bucle for, recorre las tareas de la lista e imprime cada una de ellas por pantalla. Imprime también el número de tareas que se encuentran pendientes de realizar.

Llama primero a la función añadir_tarea con una tarea como argumento para después llamar a la función gestionar_tareas para comprobar que la tarea se encuentra añadida a la lista.

```python
def añadir_tarea(tarea):
    archivo = open("44 tareas.txt", "r")
    tareas = archivo.read().splitlines()
    archivo.close() #Opcional
    
    tareas += [tarea]
    return tareas

def gestionar_tareas(tareas):
    j = 0
    print("Tareas pendientes de realizar:\n")
    for tarea in tareas:
        j += 1
        print(f"{j}. {tarea}")
    else:
        print(f"\nHay {j} tareas pendientes de realizar.")


gestionar_tareas(añadir_tarea("Peinar al gato"))
```
