**Práctica Listas**

**Objetivo**

Crea un sistema de calificaciones de estudiantes basado en métodos de objetos tipo Lista en Python.



Instrucciones

Crea una clase llamada SistemaCalificaciones que administre las calificaciones de los estudiantes en un curso.

La clase debe tener un atributo calificaciones que inicialmente sea una lista vacía. Este atributo almacenará las calificaciones de los estudiantes.

Implementa un método añadir_calificacion que reciba como parámetro calificacion y que utilice el método .append() para añadir una nueva calificación a la lista calificaciones.

Añade un método mostrar_calificaciones que imprima todas las calificaciones almacenadas mediante un bucle for.

Incluye un método calificacion_promedio que calcule y retorne el promedio de las calificaciones usando los métodos de lista apropiados (Pista: Es necesario comprobar que la lista no esté vacía. Para calcular el promedio se propone el uso de la función sum y len).

Crea un método ordenar_calificaciones que ordene las calificaciones de menor a mayor utilizando el método .sort().

Ejecución del programa:

Crea un objeto de la clase SistemaCalificaciones y llama al método añadir_calificacion con las calificaciones: 8, 9.5, 7 y 10.

Llama al método mostrar_calificaciones

Llama a la función calificacion_promedio e imprime el resultado.

Llama al método ordenar_calificaciones

Llama al método mostrar_calificaciones para comprobar que han sido ordenadas.



Resultado esperado

Calificaciones originales:

8

9.5

7

10

Promedio: 8.625

Calificaciones ordenadas:

7

8

9.5

10

```python
#SOLUCION 1

class SistemaCalificaciones:
    
    def __init__(self):
        self.calificaciones = []
    
    def añadir_calificacion(self,calificacion):
        self.calificaciones.extend(calificacion)
        
    def mostrar_calificaciones(self):
        for calificacion in self.calificaciones:
            print(calificacion,"\n")
    
    def calificacion_promedio(self):
        if self.calificaciones != []:
            return sum(self.calificaciones) / len(self.calificaciones)
        else:
            print("Lista vacía")
    
    def ordenar_calificaciones(self):
        self.calificaciones.sort()


# === MAIN ===
objeto = SistemaCalificaciones()
objeto.añadir_calificacion([8, 9.5, 7, 10])

print("Calificaciones originales:")
objeto.mostrar_calificaciones()

print("Promedio:", objeto.calificacion_promedio())
objeto.ordenar_calificaciones()
print("Calificaciones ordenadas:")
objeto.mostrar_calificaciones()
```
