````python
# SOLUCIÓN 

class Animal:
    
    def __init__(self, nombre, especie, edad, habitat):
        self._nombre = nombre
        self._especie = especie
        self._edad = edad
        self._habitat = habitat
    
    def mostrar_info(self):
        print(f"Nombre: {self._nombre}\nEspecie: {self._especie}\nEdad: {self._edad}\nHábitat: {self._habitat._tipo_habitat}, {self._habitat._temperatura}")


class Habitat:
    
    def __init__(self, tipo_habitat, temperatura):
        self._tipo_habitat = tipo_habitat
        self._temperatura = temperatura


class Mamifero(Animal):
    
    def __init__(self, nombre, especie, edad, habitat, tipo_pelaje):
        super().__init__(nombre, especie, edad, habitat)
        self._tipo_pelaje = tipo_pelaje
    
    def mostrar_info(self):
        print(super().mostrar_info(), "\nTipo de pelaje:", self._tipo_pelaje)


class Ave(Animal):
    
    def __init__(self, nombre, especie, edad, habitat, tipo_plumaje):
        super().__init__(nombre, especie, edad, habitat)
        self._tipo_plumaje = tipo_plumaje

    def mostrar_info(self):
        print(super().mostrar_info(), "\nTipo de plumaje:", self._tipo_plumaje)


class Zoologico:
     
    def __init__(self, nombre):
         self._nombre = nombre
         self._animales = []
    
    def añadir_animal(self, animal):
        self._animales.append(animal)
    
    def mostrar_animales(self):
        print(f"Animales en {self._nombre}:\n")
        for i in self._animales:
            i.mostrar_info()



zoo1 = Zoologico("Safari")

habitat1 = Habitat("Sabana", "Cálido")
habitat2 = Habitat("Bosque", "Templado")

leon = Mamifero("Simba", "León", 5, habitat1, "corto")
canario = Ave("Piolín", "Canario", 2, habitat2, "suave")

zoo1.añadir_animal(leon)
zoo1.añadir_animal(canario)
zoo1.mostrar_animales()
````

# ¿Qué me faltó entender sobre herencia?

## 1. Diferenciar entre herencia y composición

La clase `Animal` recibe un objeto `Habitat`, por lo que un hábitat **no es un Animal**, sino que **forma parte** de él. Esto se conoce como **composición**.

Por eso, para acceder a los atributos del hábitat debo hacerlo a través del objeto:

```python
self._habitat._tipo_habitat
self._habitat._temperatura
```

y no imprimir directamente:

```python
self._habitat
```

ya que eso sólo muestra la referencia al objeto.

---

## 2. `super()` sirve para llamar a los métodos de la clase padre

En las clases `Mamifero` y `Ave` intenté hacer:

```python
self.animal.mostrar_info()
```

Pero el objeto no tiene un atributo llamado `animal`.

Cuando una clase hereda de otra, los métodos de la clase padre se pueden reutilizar mediante:

```python
super().mostrar_info()
```

De esta forma no es necesario volver a escribir el mismo código.

---

## 3. Agregar un elemento a una lista

En `Zoologico` intenté agregar un animal así:

```python
self._animales += animal
```

Pero `animal` no es una lista, sino un único objeto.

La forma correcta es:

```python
self._animales.append(animal)
```

---

## 4. Recorrer una lista de objetos

En el método `mostrar_animales` hice:

```python
for i in self._animales:
    print(self._animales)
```

Esto imprimía la lista completa en cada iteración.

Lo correcto es utilizar el objeto de cada iteración:

```python
for animal in self._animales:
    animal.mostrar_info()
```

### Ejercicio 2 - Herencia + Composición

**Objetivo**

Practicar la diferencia entre herencia y composición, así como el acceso a los atributos de un objeto contenido.

**Instrucciones**

1. Crea una clase llamada `Direccion`.

2. Debe tener los atributos:

- `calle`
- `ciudad`

3. Agrega un método `mostrar_direccion()` que imprima ambos atributos.

4. Crea una clase `Persona`.

5. Debe tener los atributos:

- `nombre`
- `direccion`

> **Importante:** `direccion` debe ser un objeto de la clase `Direccion`.

6. Agrega un método `mostrar_info()` que imprima el nombre y la dirección.

7. Crea una clase `Empleado` que herede de `Persona`.

8. Agrega un nuevo atributo:

- `puesto`

9. Sobrescribe el método `mostrar_info()` utilizando:

```python
super().mostrar_info()
```

10. Después imprime el puesto del empleado.

11. Crea al menos dos empleados con direcciones diferentes y muestra su información.

.

.

**Ejemplo de salida**

```text
Nombre: Javier
Dirección: Av. Juárez, Puebla
Puesto: Analista de Datos
```

````python
class Direccion:

    def __init__(self, calle, ciudad):
        self._calle = calle
        self._ciudad = ciudad

    def mostrar_direccion(self):
        print("Dirección:", self._calle, ",", self._ciudad)

class Persona:

    def __init__(self, nombre, direccion):
        self._nombre = nombre
        self._direccion = direccion

    def mostrar_info(self):
        print("Nombre:", self._nombre)
        self._direccion.mostrar_direccion()

class Empleado(Persona):

    def __init__(self, nombre, direccion, puesto):
        super().__init__(nombre, direccion)
        self._puesto = puesto

    def mostrar_info(self):
        super().mostrar_info()
        print("Puesto:", self._puesto)


# === MAIN ===

direccion1 = Direccion("Av. Juarez", "Oaxaca")
empleado1 = Empleado("Javier", direccion1, "Programador")

direccion2 = Direccion("Av. Sn. Francisco", "México")
empleado2 = Empleado("Jonny", direccion2, "Junior")

empleado1.mostrar_info()
empleado2.mostrar_info()
````
