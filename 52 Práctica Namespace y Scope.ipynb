**Práctica Namespace y Scope**

**Objetivo**

En esta práctica se propone la "Creación de un Simulador de Ecosistema". El alumno debe desarrollar un simulador básico de ecosistema que demuestre el concepto de namespaces y scopes en Python. En este simulador, los animales interactúan en un ecosistema con recursos limitados.



Instrucciones

Define una variable global recursos_ecosistema que represente los recursos disponibles en el ecosistema, como {"agua": 1000, "alimento": 800}.

Crea una función animal_interactua() que simule cómo un animal consume o añade recursos al ecosistema. La función debe recibir tres argumentos: el tipo de animal, la cantidad de alimentos y la cantidad agua que consumirá (debe ser un parámetro opcional con el valor None por defecto).

Dentro de esta función, modifica recursos_ecosistema basándote en el tipo de animal que interactúa.

Si el animal es de tipo "herbívoro", se le restará a los recursos del ecosistema la cantidad de agua y alimento indicada por el argumento de la función. Además, se imprimirá por pantalla la frase: Un herbívoro ha consumido X de agua y X de alimento.

Si el animal es de tipo "carnívoro", se le restará a los recursos del ecosistema únicamente la cantidad de alimento indicada por el argumento de la función. Además se imprimirá por pantalla la frase: Un carnívoro ha consumido X de alimento.

Antes de realizar las operaciones anteriores, debe comprobarse que hay recursos suficientes en el ecosistema para que sean consumidos por el animal. En caso contrario, debe imprimirse por pantalla:  Recursos insuficientes en el ecosistema y debe finalizar la ejecución de la función.

Si hay recursos suficientes en el ecosistema, se imprimirá la siguiente frase por pantalla, indicando el estado actual de los recursos del ecosistema: Estado actual del ecosistema: [recursos actuales].

Implementa una función lluvia() que simule un evento de lluvia incrementando la cantidad de agua en recursos_ecosistema. Usa la palabra clave global para asegurar que la modificación afecte a la variable global recursos_ecosistema.

La cantidad de agua añadida al ecosistema viene dada por el argumento de la función.

Además, imprime por pantalla la siguiente sentencia: ¡Ha llovido! Se añadieron X unidades de agua al ecosistema.

Fuera de las funciones, desarrolla el siguiente flujo:

Inicia el programa imprimiendo por pantalla cuáles son los recursos iniciales del ecosistema: Inicio del día en el ecosistema: [recursos actuales].

Llama a la función animal_interactua en dos ocasiones, una de ellas para un animal herbívoro y otra para un carnívoro.

Llama a la función Lluvia con una cantidad de lluvia determinada.

Tras todas las llamadas a las funciones, imprime el estado del ecosistema: Fin del día en el ecosistema: [recursos actuales].



Resultado esperado

Inicio del día en el ecosistema: {'agua': 1000, 'alimento': 800}

Un herbívoro ha consumido 200 de agua y 100 de alimento.

Estado actual del ecosistema: {'agua': 800, 'alimento': 700}

Un carnívoro ha consumido 50 de alimento.

Estado actual del ecosistema: {'agua': 800, 'alimento': 650}

¡Ha llovido! Se añadieron 200 unidades de agua al ecosistema.

Fin del día en el ecosistema: {'agua': 1000, 'alimento': 650}

```python
#SOLUCION 1

recursos_ecosistema = {
    "agua": 1000,
    "alimento": 800
}

def animal_interactua(animal, alimento, agua = 0):
    global recursos_ecosistema
    
    if recursos_ecosistema["agua"] < agua or recursos_ecosistema["alimento"] < alimento:
        print("Recursos insuficientes en el ecosistema.")
        return
    
    if animal == "herviboro":
        recursos_ecosistema["agua"] -= agua
        recursos_ecosistema["alimento"] -= alimento
        
        print(f"Un hervíboro ha consumido {agua} de agua y {alimento} de alimento.")
    else:
        recursos_ecosistema["alimento"] -= alimento
        print(f"Un carnívoro ha consumido {alimento} de alimento.")
    print("Estado actual del ecosistema:", recursos_ecosistema)


def lluvia(agua_nueva):
    global recursos_ecosistema
    
    recursos_ecosistema["agua"] += agua_nueva
    print(f"¡Ha llovido! Se añadieron {agua_nueva} unidades de agua al ecosistema.")


# === MAIN ===
print(f"Inicio del día en el ecosistema: {recursos_ecosistema}.")

animal_interactua("herviboro", 100, 200)
animal_interactua("carnivoro", 20)

lluvia(50)

print(f"Fin del día en el ecosistema: {recursos_ecosistema}.")
```
