### 1. Transformaciones entre letras mayúsculas y minúsculas

```python
texto = "hOlA MuNdO"
```

El método capitalize() pone en mayúsculas la primera letra y en minúsculas el resto de letras. 

Nos da una copia de la cadena de texto original, por lo que la var original no cambia su valor.

```python
texto.capitalize()
```

```python
texto
```

```python
texto2 = texto.capitalize()
```

```python
texto2
```

También se pueden transformar todas las letras en minúsculas con el método lower(). Y con upper() las convertimos en mayúsculas.

```python
texto.lower()
```

```python
texto.upper()
```

```python
texto
```

El método title() pone en minúsculas todo menos la primera letra de las palabras.

```python
texto.title()
```

El método swap() cambia mayúsculas por minúsculas y minúsculas por mayúsculas.

```python
texto.swapcase()
```

```python
"COLOR" == "color"
```

### 2. Buscar y reemplazar 

```python
texto = "aoo boo coo"
```

El método .count() cuenta las veces que se repite una subcadena

```python
texto.count("oo")
```

**El método .find() devuelve el índice donde se encuentra la subcadena**

```python
texto.find("boo")
```

```python
texto[4]
```

```python
texto[4:4 + len('boo')]
```

El método busca la subcadena empezando por el final

```python
texto = "aoo boo coo boo"
```

```python
texto.find("boo")
```

```python
texto.rfind("boo")
```

.rfind() busca la subadena desde la derecha de la cadena hasta la izquierda. Y el método .find() va de izquierda a derecha.

.rfind() = reverse find

El método .join() concatena un string con un objeto iterable.

```python
",".join(["azul","verde","rojo"])
```

```python
"-".join(["azul","verde","rojo"])
```

```python
print(",".join(["azul","verde","rojo"]))
```

El método .format() nos permite formatear un string para incluir variables dentro del string. Incluye los argumentos en donde nosotros hayamos puesto los corchetes ({}).

```python
"{}-{}-{}".format(3,"hola",[2,3])
```
