**Práctica Tuplas**

**Objetivo**

Programa un "Gestor de Información Personal" utilizando Tuplas en Python.



Instrucciones

El programa tendrá una función llamada gestionar_informacion que recibe una tupla datos_personales como argumento. Esta tupla contendrá tres elementos: nombre, edad y ciudad de residencia.

Dentro de la función, "desempaqueta" (unpack) la tupla datos_personales en tres variables: nombre, edad, ciudad.

Calcula el año de nacimiento de la persona basándote en su edad actual y guarda este valor en una variable nacimiento.

Crea una nueva tupla llamada datos_modificados que será una versión modificada de datos_personales. Esta tupla contendrá los siguientes elementos: nombre, nacimiento y ciudad de residencia.

La función debe retornar ambas tuplas, datos_personales y datos_modificados.

Fuera de la función, imprime las dos tuplas para mostrar la diferencia entre la tupla original y la modificada.



Resultado esperado

Tupla original: ('Ana', 30, 'Madrid')

Tupla modificada: ('Ana', 1994, 'Madrid')

```python
def gestionar_informacion(datos_personales):
    nombre, edad, ciudad = datos_personales
    nacimiento = 2025 - edad
    datos_modificados = (nombre, nacimiento, ciudad)
    return datos_personales, datos_modificados


datos_personales = ("Goku", 30, "Namek")

tupla_original, tupla_modificada = gestionar_informacion(datos_personales)

print("Tupla original:", tupla_original)
print("Tupla modificada:", tupla_modificada)
```
