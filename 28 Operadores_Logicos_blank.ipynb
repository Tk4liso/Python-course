### 1. ¿Cuáles son los operadores lógicos?

### 2. Operador `not`

```python
num = 5
```

```python
num < 10
```

```python
not num < 10
```

```python
not (num < 10)
```

```python
num = 5
print(not num)
```

```python
num = -5
print(not num)
```

```python
num = 0
print(not num)
```

```python
num = 1
print(not num)
```

Si es **cero = False**, por lo tanto not False = True.

Si es **diferente de cero (pos/neg) = True**, por lo tanto, not True = False.

### 3. Operador `or`

```python
num1 = 5 
num2 = 10
```

```python
num1 < 4
```

```python
(num1 < 4) or (num2 > 15) or (num1 < 10)
```

### 4. Operador `and`

```python
num1 = 5
num2 = 10
```

```python
num1 < 3 and num2 > 5 and num2 > 2
```

```python
(num1 < 5 or num2 > 3) and num2 == 10 and num1 < 8
```

```python
True and num2 == 10 and num1 < 8
```

```python
True and True and num1 < 8
```

```python
True and True and True
```
